```python
# This script reads a csv file containing run length encoded segmentation labels,
# creates the encoded binary images and saves them 

# This mounts your Google Drive to the Colab VM.
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# Path to images
input_path = '/content/drive/MyDrive/Deep_learning/project/data/train_semi_supervised/'

# Path to save patches
patch_path ='/content/drive/MyDrive/Deep_learning/project/data/train_semi_supervised_patches/'
```

```python
import os
import sys
import numpy as np
import cv2

for f in os.listdir(input_path):
    if f.find(".png") != -1:
        img = cv2.imread("{}/{}".format(input_path, f),0)
        file_name = f[:f.find(".png")]

        for i in range(3):
            for j in range(4):
                cv2.imwrite(f"{patch_path}{file_name}{str(i)}{str(j)}.png",img[148*i:224+148*i, 160*j:224+160*j])
                        
```
