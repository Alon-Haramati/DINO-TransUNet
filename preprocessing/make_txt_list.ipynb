```python
with open(list_path + "train.txt", 'w') as train_output:
    with open(list_path + "test.txt", 'w') as test_output:

        for f in os.listdir(train_path):
            if f.find(".npz") != -1:

              file_name = f[:f.find(".npz")]
              train_output.write(file_name+'\n')

        for f in os.listdir(test_path):
            if f.find(".npz") != -1:

              file_name = f[:f.find(".npz")]
              test_output.write(file_name+'\n')

      



```
