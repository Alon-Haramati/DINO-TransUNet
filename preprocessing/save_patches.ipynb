```python
# This script reads a csv file containing run length encoded segmentation labels,
# creates the encoded binary images and saves them 

# This mounts your Google Drive to the Colab VM.
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# Path to train images
input_path = '/content/drive/MyDrive/Deep_learning/project/data/train/'
label_path = '/content/drive/MyDrive/Deep_learning/project/data/labels/'

# Path to save .npz files
train_path ='/content/drive/MyDrive/Deep_learning/project/data/train_npz_patches/'
test_path ='/content/drive/MyDrive/Deep_learning/project/data/test_npz_patches/'

# Path to save image list
list_path ='/content/drive/MyDrive/Deep_learning/project/TransUnet_copy/lists/lists_Sartorius_patches/'
```

```python
import os
import sys
import numpy as np
import cv2

class Utils:
    @staticmethod
    def get_preprocessed_img(img_path,label=False):
        img = cv2.imread(img_path,0)
        img = img.astype(np.float32)
        # Normalize to range [0 1]
        img = (img-np.min(img))/(np.max(img)-np.min(img))
        if label:
          img = 1- img
        return img

print("Converting images to numpy arrays...")

with open(list_path + "train.txt", 'w') as train_output:
    with open(list_path + "test.txt", 'w') as test_output:

        ind = 1
        for f in os.listdir(input_path):
            if f.find(".png") != -1:
                img = Utils.get_preprocessed_img("{}/{}".format(input_path, f))
                label = Utils.get_preprocessed_img("{}/{}".format(label_path, f),label=True)
                file_name = f[:f.find(".png")]

                for i in range(3):
                    for j in range(4):
                        if ind<=500:
                            np.savez_compressed("{}/{}".format(train_path, file_name+str(i)+str(j)),
                                                image=img[148*i:224+148*i, 160*j:224+160*j],
                                                label=label[148*i:224+148*i, 160*j:224+160*j])
                            train_output.write(file_name+str(i)+str(j) + '\n')
                        else:
                            np.savez_compressed("{}/{}".format(test_path, file_name+str(i)+str(j)),
                                                image=img[148*i:224+148*i,160*j:224+160*j],
                                                label=label[148*i:224+148*i,160*j:224+160*j])
                            test_output.write(file_name+str(i)+str(j) + '\n')

                ind += 1


print("Numpy arrays saved in {}".format(train_path))
```
